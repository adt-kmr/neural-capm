```python
import yfinance as yf 
```

```python
stock = yf.download("RELIANCE.NS", start="2015-01-01", end="2024-12-31")
index = yf.download("^NSEI", start = "2015-01-01", end="2024-12-31")
```

```python
stock.head()
```

```python
import pandas as pd

stock_ret = stock["Close"].pct_change().dropna()
index_ret = index["Close"].pct_change().dropna()

df = pd.concat([stock_ret,index_ret], axis=1, join="inner")
df.columns = ["stock_ret", "index_ret"]
df.head()
```

```python
import numpy as np

import statsmodels.api as sm

cov = np.cov(df["stock_ret"], df["index_ret"])[0][1]
var = np.var(df["index_ret"])
beta_manual = cov/var

print("Beta (Covariance Method):", beta_manual)

```

```python
X = sm.add_constant(df["index_ret"])
model = sm.OLS(df["stock_ret"], X).fit()
print(model.summary())
print("Beta (OLS): ", model.params["index_ret"])

```
